# Relaciones entre datos

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from sklearn import linear_model
import pylab
from pylab import rcParams
import seaborn as sns
%matplotlib inline
rcParams['figure.figsize'] = 10, 10
```

```python
defunciones = pd.read_csv('pablo_reduced.csv', sep=';', decimal = '.')
defunciones = defunciones.rename(index=str, columns={"Unnamed: 0": "Year"})
defunciones_grouped = defunciones.groupby(['Year']).mean()
defunciones_grouped.head(5)
```

```python
atomosferico = pd.read_csv('curro_reduced.csv', sep=';', decimal = '.')
atomosferico_grouped = atomosferico.groupby(['Year']).mean()
atomosferico_grouped.head(5)
```

```python
frames = [atomosferico_grouped, defunciones_grouped]
result = pd.concat(frames, axis=1)

result.head(5)
```

```python
plt.plot(result)
plt.legend(result)
```

```python
result.columns
```

```python
altas = pd.read_csv('altas.csv', sep=';', decimal = '.')
altas = altas.rename(index=str, columns={"Año": "Year"})
altas_grouped = altas.groupby(['Year']).mean()
altas_grouped.head(5)
```

```python
camas = pd.read_csv('camas.csv', sep=';', decimal = '.')
camas = camas.rename(index=str, columns={"Año": "Year"})
camas_grouped = camas.groupby(['Year']).mean()
camas_grouped.head(5)
```

```python
estancias = pd.read_csv('estancias.csv', sep=';', decimal = '.')
estancias = estancias.rename(index=str, columns={"Año": "Year"})
estancias_grouped = estancias.groupby(['Year']).mean()
estancias_grouped.head(5)
```

### Cuanto más ozono, menos radiación

```python
ax = sns.regplot(result['Ozono'].values, result['TotalRadiacion'].values)
ax.set(xlabel = 'Ozono', ylabel = 'Radiación')
```

### Cuanto más temperatura, menos precipitaciones

```python
bx = sns.regplot(result['Temperatura media mensual (ºC)'].values, result['Precipitaciones / Total (mm.)'].values)
bx.set(xlabel = 'Temperatura', ylabel = 'Precipitaciones')
```

### Cuanto más precipitaciones, menos partículas en el aire

```python
cx = sns.regplot( result['Particulas suspension  (gr/m3)'].values, result['Precipitaciones / Total (mm.)'].values)
cx.set(ylabel = 'Precipitaciones', xlabel = 'Partículas aire')
```

### Cuanto más ingresdos en hospital por psiquiatría, más fallecidos pos trastornos mentales

```python
dx = sns.regplot( result['trastornos_mentales'].values, estancias['Psiquiatría'].values)
dx.set(xlabel = 'Defunciones por trastornos mentales', ylabel = 'Ingresados en el hospital por psiquiatría')
```

### Cuanto más ingresdos en hospital de larga estancia, mas fallecidos por tumores

```python
ex = sns.regplot( result['tumores'].values, estancias['Larga Estancia'].values)
ex.set(xlabel = 'Defunciones por tumores', ylabel = 'Ingresados en el hospital de larga estancia')
```

### Cuantas más estancias en hospital, más camas

```python
fx = sns.regplot( camas['Total año'].values, estancias['TOTAL'].values)
fx.set(xlabel = 'Camas de hospital', ylabel = 'Estancias en hospital')
```
